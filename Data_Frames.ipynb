```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
  
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
result = left.join(right, how='outer', lsuffix='_left', rsuffix='_right')
```

```python
newdf = result[(result['A'] > 'A2') & (result['C'] == 'C2')]
newdf
```

```python
pivot_table = result.pivot_table(values='D', index=['A', 'B'], columns=['C'])
pivot_table
```
